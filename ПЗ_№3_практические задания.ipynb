# Площадь под гиперболой
Найдите площадь под функцией:
$f(x) =\frac{1}{x}$
на интервале [1,5].

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(1, 5, 50)
y = 1/x

plt.figure(figsize=(12, 8))
plt.plot(x, y)

plt.fill_between(x, y, where=(y > 0))

integral_value = np.trapezoid(y, x)
print(f"Площадь равна: {integral_value}")
```

# Коническая поверхность
Постройте конус:
$f(x, y) = \sqrt{x^2 + y^2}$

на интервале х, у ∈ [-5, 5].

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)


fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis, alpha=0.8)
ax.plot_wireframe(X, Y, Z, color='black', alpha=0.3, linewidth=0.5)
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='Значение функции')


ax.set_xlabel('Ось X')
ax.set_ylabel('Ось Y')
ax.set_zlabel('Ось Z')
ax.set_title('График функции $f(x, y) = \sqrt{x^2 + y^2}$', fontsize=14)
```
