# 1. Построение простого графика
Напишите код, который строит график функции $y = x^2$ на интервале [−10, 10].
Добавьте заголовок, подписи осей и сетку.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(-10, 10, 50)
y = [i**2 for i in x]
plt.figure(figsize=(12, 8))
plt.title("$y2 = x^2$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.plot(x, y)

```

# 2. Настройка стилей и цветов
Постройте три линии на одном графике:

*   $y = x $(синяя, пунктирная линия),
*   $у = x2$ (зеленая, штрих-пунктирная линия),
*   $y = x3$ (красная, сплошная линия).

Добавьте легенду и сделайте оси одинакового масштаба.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(-10, 10, 100)
y = x
y1 = x**2
y2 = x**3

plt.figure(figsize=(12, 12))
plt.title("$y2 = x^2$")
plt.xlabel("x")
plt.ylabel("y")

plt.axis('equal')
plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.legend()
plt.grid()
plt.plot(x, y,  label="$y=x$", color="blue", linestyle="dotted")
plt.plot(x, y1, label="$у = x2$", color="green", linestyle="dashdot")
plt.plot(x, y2, label="$y = x3$", color="red", linestyle="solid")
plt.legend()
```

#7. Трехмерный график поверхности
Используя mpl_toolkits.mplot3d , постройте 3D-график функции $z = sin(\sqrt{x^2 + y^2)}$ на сетке значений x, у в диапазоне [-5, 5].

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)


Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D-график функции $z = \sin(\sqrt{x^2 + y^2})$')

plt.show()
```

# 8. Множественные подграфики (subplots)
Постройте четыре графика в одной фигуре (2 × 2 сетка):


1.   Линейный график $у = x$
2. Парабола $у =2$
3. Синус $у = sin(x)$
4. Косинус $у = cos(x)$

Добавьте заголовки к каждому подграфику.

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-5, 5, 100)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].plot(x, x,  linewidth=2)
axes[0, 0].set_title('$y = x$')
axes[0, 0].grid(True)


axes[0, 1].plot(x, x**2,  linewidth=2)
axes[0, 1].set_title('$y = x^2$')
axes[0, 1].grid(True)


axes[1, 0].plot(x, np.sin(x),  linewidth=2)
axes[1, 0].set_title('$y = sin(x)$')
axes[1, 0].grid(True)

axes[1, 1].plot(x, np.cos(x),  linewidth=2)
axes[1, 1].set_title('$y = cos(x)$')
axes[1, 1].grid(True)
```
